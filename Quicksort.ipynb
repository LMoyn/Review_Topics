```python
import math
import os
import random
import re
import sys

#quicksort with last element as pivot

def quicksort(arr,low,high):
    if(low < high):
        #partitions the array and recursively quicksorts each half
        partitioning_index = partition(arr,low,high)
        quicksort(arr,low,partitioning_index -1)
        quicksort(arr,partitioning_index +1,high)

def partition(arr,low,high):
    #Takes the last element of the array as the pivot
    pivot = arr[high]
    
    
    i = low -1

    #Iterates through the array
    for j in range (low,high):
        #swaps the current element with a smaller one if it is found
        if arr[j] < pivot:
            i = i+1
            temp = arr[i]
            arr[i] = arr[j]
            arr[j] = temp

    #Swaps the value at the end of the array with the i+1 value
    temp = arr[high]
    arr[high] = arr[i+1]
    arr[i+1] = temp
    #Returns the index i+1
    return i + 1


#Test array for illustration
array = [ 11, 17, 8, 19, 1, 51 ]
print(array)
quicksort(array,0, len(array) - 1)
  
print(f'Sorted array: {array}')
```
