```python
def bubble_sort(arr):
    #Makes sure to perform inner loop n times
    for i in range (0,len(arr)):
        #iterates through all elements of array, swapping the current one and one after it if they are in the wrong order
        for j in range (0,len(arr)-1):
            if arr[j] > arr[j+1]:
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
                
#Test array for illustration
array = [ 11, 17, 8, 19, 1, 51 ]
print(array)
bubble_sort(array)
  
print(f'Sorted array: {array}')
```
